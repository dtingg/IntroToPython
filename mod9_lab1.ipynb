```python
#import modules
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
#run plots in the notebook
%matplotlib inline 
url = "http://pbpython.com/extras/sample-salesv2.csv" 
sales = pd.read_csv(url)
```

```python
#Rename the columns to use underscores instead of spaces
sales.columns = ['account_number', 'name', 'sku', 'category', 'quantity', 'unit_price',
                 
#Subset the dataframe to contain only shirt sales
       'ext_price', 'date']
```

```python
#Subset the dataframe to contain only the name, category, quantity and unit price columns
sales[["name", "category", "quantity", "unit_price"]]
```

```python
#Subset the dataframe to contain only shirt sales
shirt_sales = sales[sales["category"] == "Shirt"]
shirt_sales
```

```python
# Calculate the total cost per shirt sale
shirt_sales["shirt_sales"] = shirt_sales.quantity * shirt_sales.unit_price

```

```python
shirt_sales
```

```python
# Group the shirt sales by company name
shirts_by_company = shirt_sales.groupby("name").sum()

shirts_by_company
```

```python
#Pull out top 10 shirt sales
top_sellers = shirts_by_company.sort_values(by="shirt_sales", ascending=False).head(10)
top_sellers = top_sellers.reset_index()
top_sellers
```

```python
# Graph the top 10 shirt sales
shirt_plot = top_sellers.plot(kind="bar", title="Total Shirt Sales by Company", x="name", y="shirt_sales")
shirt_plot.set_xlabel("Company Name")
shirt_plot.set_ylabel("Shirts Sold ($)")
```
